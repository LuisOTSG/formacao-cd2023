## Pandas

```python
# Importar a biblioteca pandas
import pandas as pd
```

```python
# Carregar arquivo para dataframe Pandas
dados = pd.read_csv('../dados/Credit.csv') 
# Primeiros registros
dados.head()
```

```python
# Verificar formato (quantidade de linhas e de colunas)
dados.shape
```

```python
# Resumo estatístico das colunas numéricas
dados.describe()
```

```python
# Últimos registros
dados.tail()
```

```python
# Retornar o nome das colunas
dados.columns
```

```python
# Filtrar por nome da coluna
dados[["duration"]]
```

```python
# Filtrar linhas por índice, todas as colunas
dados.loc[1:3]
```

```python
# Filtrar linhas 1 e 3, todas as colunas
dados.loc[[1,3]]
```

```python
# Filtrar coluna por registros específicos
dados.loc[dados['purpose'] == "radio/tv"]
```

```python
# Filtrar coluna por registros específicos
dados.loc[dados['credit_amount'] > 15000]
```

```python
# Aribuir resultado a variável, criando outro df
credito2 = dados.loc[dados['credit_amount'] >  18000]
credito2.head()
```

```python
# Definir só algumas colunas
credito3 = dados[['checking_status','duration']].loc[dados['credit_amount'] >  18000]
credito3.head()
```

```python
# Series (unica coluna)
# Criar Series a partir de lista
s1 = pd.Series([2,5,3,34,54,23,1,16])
print(s1)
print(type(s1))
```

```python
# Criar Series a partir de um array do numpy
import numpy as np
array1 = np.array([2, 5, 3, 34, 54, 23, 1, 16])
s2 = pd.Series(array1)
print(s2)
print(type(s2))
```

```python
# Criar Series a partir de um dataframe
s3 = dados['purpose']
print(s3)
type(s3)
```

```python
# Temos um dataframe a partir de uma coluna
d4 = dados[['purpose']]
print(d4)
type(d4)
```

```python
# Renomear colunas (sem persistir, ou seja, sem substituir o dataframe)
dados.rename(columns={"duration": "duração","purpose": "propósito"})
```

```python
dados.head()
```

```python
# Renomear colunas (persistindo, ou seja, substituindo o dataframe)
dados.rename(columns={"duration": "duração","purpose": "propósito"}, inplace=True)
dados.head()
```

```python
# Excluir coluna (sem persistir)
dados.drop('checking_status', axis=1)
```

```python
dados.head()
```

```python
# Excluir coluna (persistindo)
dados.drop('checking_status', axis=1, inplace=True)
dados.head()
```

```python
# Verificar dados nulos/ausentes
dados.isnull().sum()
```

```python
# Retirar colunas com NaN
dados.dropna()
```

```python
# Preencher dados faltantes
dados['duração'].fillna(0, inplace = True)
dados.head()
```

```python
# iloc() - Filtra apenas a partir do índice
dados.iloc[0:3,0:5]
```

```python
dados.iloc[[0, 1, 2, 3, 7], 0:5]
```
